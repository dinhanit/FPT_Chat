```python
def remove_txt(sentence, file):
    # Read the stopword file and store the words in a set
    with open(file, 'r', encoding='utf8') as f:
        file = set(word.strip() for word in f.readlines())
    # Split the sentence into individual words
    words = sentence.split()
    # Check if each word is a stopword or not
    clean_words = [word for word in words if word.lower() not in file]
    # Join the words back together to form the updated sentence
    clean_sentence = ' '.join(clean_words)

    return clean_sentence

file = "stopword.txt"  # Replace with the path to your file
```
