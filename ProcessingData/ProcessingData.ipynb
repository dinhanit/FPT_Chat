```python
def show_stopwords(sentence, stopword_file): #Show
    with open(stopword_file, 'r', encoding='utf-8') as f:
        stopwords = set(word.strip() for word in f.readlines())
    
    words = sentence.split()
    stop_words = [word for word in words if word.lower() in stopwords]
    if stop_words:
    # Join the words back together to form the updated sentence
        return sentence
    else:
        return None

stopword_file = "pro_noun.txt"

# print(show_stopwords('ai chua ong nay rang hieu y cua e',stopword_file))
# print(show_stopwords('bac lam cung',stopword_file))
```

```python
def remove_pro_noun(sentence, pronoun_file):
    # Read the stopword file and store the words in a set
    with open(pronoun_file, 'r', encoding='utf8') as f:
        pronoun_file = set(word.strip() for word in f.readlines())
    # Split the sentence into individual words
    words = sentence.split()
    # Check if each word is a stopword or not
    clean_words = [word for word in words if word.lower() not in pronoun_file]
    # Join the words back together to form the updated sentence
    clean_sentence = ' '.join(clean_words)

    return clean_sentence
pronoun_file = "pro_noun.txt"  # Replace with the path to your stopword file
```

```python
def translate_emoticons_to_Vietnamese(sentence, translate_emoticon_file):
    with open(translate_emoticon_file, 'r', encoding='utf8') as file:
        dictionary = json.load(file)
    # Split the sentence into individual words
    word_list = sentence.split()
    for i in range(len(word_list)):
        if word_list[i] in dictionary:
            word_list[i] = dictionary[word_list[i]]
     # Join the words back together to form the updated sentence        
    modified_sentence = ' '.join(word_list)

    return modified_sentence
translate_emoticon_file = "translate_emoticons.json"

column_data = df_new['Feedback']
for value in column_data:
    print(translate_emoticons_to_Vietnamese(value,translate_emoticon_file))

```
